```python
sns.boxplot(y = 'Landslide',data= df, x= 'Precipitation')
plt.xlabel('Precipitation')
```

 The above boxplot shows when the precipitation is above 3 there is a chance of Landslide.


#

```python
sns.histplot(df['Curvature'], bins=15, kde=True, color ="green")

plt.title("Distribution of Curvature ")
plt.show()
```

 The above histogram shows that the highest distribution curvature is 3.0.


#

```python
sns.barplot(data=df, x='NDWI', y='Landslide', palette='coolwarm')

# Customize the plot
plt.title('Landslide on the basis of NDWI')
plt.xlabel('NDWI')
plt.ylabel('Landslide')

# Show the plot
plt.show()
```

 The above figure shows the possibility of landslide on the basis of NDWI (Normalized Difference Water Index)


#

```python
sns.violinplot(data=df, x='Slope', y='Landslide', palette='viridis')
plt.title('Landslide probability according to slope')
plt.xlabel('Slope')
plt.ylabel('Landslide')
plt.show()
```

 The above violin plot shows that the probability of occurance of ladnslide on the basis of the slope.
